# Basic Import

```python
import matplotlib.pyplot as plt
```

# Basic Plotting

```python
# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
```

```python
plt.plot(x,y)
plt.title(' Line Plot ')
plt.xlabel('X values')
plt.ylabel('y values')
plt.show()
```

# Customisation and Styling

```python
plt.plot(x, y, color='red',marker='o',linestyle='--',linewidth=2,markersize=8)
plt.title(' Styled Plot', fontsize=16)
plt.xlabel("X-axis",fontsize=12)
plt.ylabel("Y-axis",fontsize=12)
plt.grid(True, linestyle=':',linewidth=0.5)
plt.show()
```

# Types of plots

## Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories,values,color='blue')
plt.title('Bar Plot',fontsize=16)
plt.xlabel('Categories',fontsize=12)
plt.ylabel('Values',fontsize=12)
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories,values,color=['r','b','g','y'])
plt.title('Bar Plot',fontsize=16)
plt.xlabel('Categories',fontsize=12)
plt.ylabel('Values',fontsize=12)
plt.legend()
plt.show()
```

## Histogram

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=50,color='orange',edgecolor='black')
plt.title('Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

## Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(50, 300, size=50)
colors = np.random.rand(50)

plt.scatter(x,y, s=sizes, c=colors, alpha=0.5,cmap='viridis') # cmap='viridis' ensures a smooth gradient of colors.
plt.colorbar() # legend
plt.title('Scatter Plot')
plt.show()
```

## Pie Chart

```python
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True, startangle=140)
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```

## Box Plot

```python
data = [np.random.rand(50), np.random.rand(50)]

plt.boxplot(data, labels=['Group1','Group2'])
plt.title("Box Plot")
plt.ylabel('Values')
plt.show()
```
